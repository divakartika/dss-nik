# Data Customer

```python
kodewilayah = pd.read_csv('data/kodewilayah.csv', header=None)
kodewilayah.rename(columns={0:'kode', 1:'wilayah'}, inplace=True)


kodekec = kodewilayah[kodewilayah['kode'].str.len() == 8].copy()
kodekec['kode'] = kodekec['kode'].str.replace('.', '')

customer = pd.read_csv('data/customer.csv')
```

```python
customer = customer[customer['Age'].between(17,68)].reset_index(drop=True)
customer['Profession'].fillna('Not Employed', inplace=True)

size = customer.shape[0]
size
```

```python
customer.describe()
```

```python
customer.isna().sum()
```

```python
employ_all = employee[employee['STATUS'] == 'ACTIVE'].copy()
employ_2k = employ_all.sample(ignore_index=True, n=size).copy()
employ_2k
```

```python
from datetime import date
customer['birth_year'] = date.today().year - customer['Age']
customer.head()
```

```python
customer.isna().sum()
```

```python
employ_2k['birthdate_dtm'] = pd.to_datetime(employ_2k['birthdate_key'])
customer['birthdate_str'] = employ_2k['birthdate_dtm'].dt.strftime('%d%m') + customer['birth_year'].astype('str').str[-2:]
customer.isna().sum()
```

```python
customer['birthdate_int'] = customer['birthdate_str'].astype('int64')
customer.head()
```

```python
customer['birthdate_int'] += (customer['Gender'] == 'Female') * 400000
customer['birthdate_str'] = customer['birthdate_int'].astype('str').str.zfill(6)
```

```python
customer.head()
```

```python
customer['kodekec'] = np.random.choice(kodekec['kode'].unique(), size)
customer['kodekec']

customer['reg'] = np.random.randint(low=1, high=20, size=size).astype('str')
customer['reg'] = customer['reg'].str.zfill(4)

customer['NIK'] = customer['kodekec'] + customer['birthdate_str'] + customer['reg']

cols = ['Gender', 'Age', 'birth_year', 'birthdate_str', 'birthdate_int', 'kodekec', 'reg']

customer_clean = customer.drop(columns=cols).reset_index(drop=True)

customer_clean = customer_clean.iloc[:,[0,-1,1,2,3,4,5]]
customer_clean.rename(columns={"Annual Income ($)": "Annual_Income", 
                         "Spending Score (1-100)": "Spending_Score", 
                         "Work Experience": "Work_Experience",
                         "Family Size": "Family_Size"}, inplace=True)

customer_clean['Annual_Income'] = customer_clean['Annual_Income'] * 1000

customer_clean.head()
```
